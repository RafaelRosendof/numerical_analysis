```python
###### Atividade 


import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(-10, -1, 0.01)
y1 = x1 + 1

x2 = np.arange(-1, 2, 0.01)
y2 = -x2**2 + 2

x3 = np.arange(2, 10, 0.01)
y3 = x3

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x1, y1, label='f(x) = x + 1', color='blue')
ax.plot(x2, y2, label='f(x) = -x^2 + 2', color='green')
ax.plot(x3, y3, label='f(x) = x', color='red')

# 
ax.plot([-1, -1], [-10, 2], linestyle='--', color='black')
ax.plot([2, 2], [-10, 10], linestyle='--', color='black')

# 
ax.plot([-1], [0], marker='o', markersize=7, color="black")
ax.plot([2], [1], marker='o', markersize=7, color="black")


ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Gráfico das Funções')

plt.grid(True)
plt.show()
```
